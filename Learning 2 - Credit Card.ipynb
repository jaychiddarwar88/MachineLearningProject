```python
import keras
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score as accsc
from sklearn.impute import SimpleImputer as Imp
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder 
import time
```

```python
df = pd.read_csv("BankChurners.csv")
```

```python
df = df.iloc[:, :21]
```

```python
y = df['Attrition_Flag'].values
x = df.drop(['Attrition_Flag', 'CLIENTNUM' ], axis=1)
```

```python
labely = LabelEncoder()
```

```python
y = labely.fit_transform(y)
```

```python
labelx = LabelEncoder()
labels = ['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']  
for label in labels:
    labelenc = LabelEncoder()
    x[label] = labelenc.fit_transform(x[label])

```

```python
xtrain, xtest, ytrain, ytest = tts(x, y, test_size = 0.3, random_state = 8)
```

```python
scaler = StandardScaler()
```

```python
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(19, input_dim = 19, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss= 'mae', optimizer='adam', metrics=['accuracy'])
```

```python
start = time.time()
model.fit(xtrain, ytrain, epochs=100, batch_size = 20)
```

```python
accuracy = model.evaluate(xtest, ytest)
end = time.time()
print(end - start)
print(accuracy)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'linear' )
svc.fit(xtrain, ytrain)
ypredsvc = svc.predict(xtest)

print(accsc(ytest, ypredsvc))
```

```python

from sklearn.neighbors import KNeighborsClassifier as KNN

knn = KNN(n_neighbors = 11, metric = 'euclidean')
knn.fit(xtrain, ytrain)
ypredknn = knn.predict(xtest)

print(accsc(ytest, ypredknn))
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)
clf.fit(xtrain,ytrain)

ypredrfc = clf.predict(xtest)

print(accsc(ytest, ypredrfc))
```
