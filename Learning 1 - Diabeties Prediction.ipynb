```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score as accsc
from sklearn.impute import SimpleImputer as Imp
```

```python
df = pd.read_csv("Datasetdia.csv")
```

```python
df.columns = ['times preg', 'plsglucose', 'blood pressure', 'triceps skin fold', '2 hrs insulin', 'bmi', 'dia pedigree', 'age', 'pred']
```

```python
# missing values
df[['plsglucose', 'blood pressure', 'triceps skin fold', '2 hrs insulin' , 'bmi']] = df[['plsglucose', 'blood pressure', 'triceps skin fold', '2 hrs insulin' , 'bmi']].replace(0, np.NaN)

```

```python
df.head()
```

```python
x = df.iloc[:, :8].values
y = df.iloc[:, 8].values

imputer = Imp(np.nan , 'mean' , 0)
imputer = imputer.fit(x[:,1:6])
x[:,1:6]= imputer.transform(x[:,1:6])
```

```python
xtrain, xtest, ytrain, ytest = tts(x, y , test_size = 0.3, random_state= 8)

scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from pickle import dump

dump(scaler, open('diabetiessc.pkl', 'wb'))
```

```python
# Artificial Neural Network

from keras.models import Sequential
from keras.layers import Dense, Dropout
import keras

model = Sequential()
model.add(Dense(8 , input_dim=8 , activation='tanh'))
model.add(Dropout(0.55))
model.add(Dense(8, activation='tanh'))
model.add(Dropout(0.55))
model.add(Dense(8, activation='tanh'))
model.add(Dropout(0.55))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mae', optimizer='adam', metrics=['accuracy'])

#es_callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
model.fit(xtrain, ytrain, epochs=500, batch_size=20)

accuracy = model.evaluate(xtest, ytest)
print("Accuracy of ANN is : " +  str(float(accuracy[1] * 100)))
print("Loss of ANN is : " +  str(float(accuracy[0])))
```

```python
# Support Vector Classifier

from sklearn.svm import SVC

svc = SVC(kernel = 'linear' )
svc.fit(xtrain, ytrain)
ypredsvc = svc.predict(xtest)

print(accsc(ytest, ypredsvc))
```
